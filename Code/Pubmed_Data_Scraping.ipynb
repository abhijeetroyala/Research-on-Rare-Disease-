```python
from Bio import Entrez

def search(query):
    Entrez.email = 'farheennz93@gmail.com'
    handle = Entrez.esearch(db='pubmed', 
                            sort='relevance', 
                            retmax='6763',
                            retmode='xml', 
                            term=query)
    results = Entrez.read(handle)
    return results

def fetch_details(id_list):
    ids = ','.join(id_list)
    Entrez.email = 'farheennz93@gmail.com'
    handle = Entrez.efetch(db='pubmed',
                           retmode='xml',
                           id=ids)
    results = Entrez.read(handle)
    return results

if __name__ == '__main__':
    results = search('sjogren') #here you can replace with another disease name like search('astama') etc.
    id_list = results['IdList']
    papers = fetch_details(id_list)
    for i, paper in enumerate(papers['PubmedArticle']):
        print("%d) %s" % (i+1, paper['MedlineCitation']['Article']['ArticleTitle']))
    
```

```python
#Scrapping the Titles

titles=[]
for i, paper in enumerate(papers['PubmedArticle']):
        q=(paper['MedlineCitation']['Article']['ArticleTitle'])
        titles.append(q)
titles
```

```python
len(titles)
```

```python
#Saving titels into file(contains only titles)
with open('sjogren_pubmed.txt', 'w',encoding="utf-8") as f:
    for item in titles:
        f.write("%s\n" % item)
```

```python
#Scrapping the Abstracts

import json
abstracts=[]
for i in range(6753):
    try:
        a=json.dumps(str(papers['PubmedArticle'][i]['MedlineCitation']['Article']['Abstract']['AbstractText']).strip('[]'), indent=2, separators=(',', ':'))
        abstracts.append(a)
    except:
        pass
abstracts
```

```python
len(abstracts)
```

```python
#Tokenizing the Abstracts

from nltk import tokenize
b=[]
for i in range(len(abstracts)):
    a=tokenize.sent_tokenize(abstracts[i])
    b.append(a)
b
```

```python
len(b)
```

```python
c=sum(b, [])
```

```python
len(c)
```

```python
sentences=c
```

```python
len(c)
```

```python
#Saving into file(contains only abstracts no titles)
with open('sjogren_pubmed_abs.txt', 'w',encoding="utf-8") as f:
    for item in sentences:
        f.write("%s\n" % item)
```
