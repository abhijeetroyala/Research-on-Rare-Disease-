```python
with open("sjogren_pubmed_abs.txt", 'r',encoding="utf-8") as f:
    sjogren = [line.lower().rstrip('\n') for line in f]
len(sjogren)
```

```python
sens1=[i for i in sjogren if 'stringelement' not in i.strip().lower()]
sens2=[i for i in sens1 if 'attributes=' not in i.strip().lower()]
```

```python
sens2
```

```python
search_keywords=["sjogren","sj\\u00f6gren"]
```

```python
sjogren_sentences=[]
for sentence in sens2:
    if (any(map(lambda word: word in sentence, search_keywords))):
        sjogren_sentences.append(sentence)
len(sjogren_sentences)
```

```python
sjogren_sentences
```

```python
a=[]
for i in sjogren_sentences:
    b=i.replace("sj\\u00f6gren","sjogren")
    a.append(b)
```

```python
a
```

```python
len(a)
```

```python
with open('sjogren_dataset.txt', 'w',encoding="utf-8") as f:
    for item in a:
        f.write("%s\n" % item)
```
